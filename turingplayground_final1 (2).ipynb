```python
class StrategyFeatureGenerator:
    def __init__(self, model, feature_names):
        self.model = model
        self.feature_names = feature_names

# in this class first it make imp_df in which feature name and their importance

    def get_importance_quantiles(self):
        imp = self.model.feature_importances_
        imp_df = pd.DataFrame({
            "feature": self.feature_names,
            "importance": imp
        }).sort_values("importance", ascending=False)

        imp_df["quantile"] = 0.1 + 0.7 * ((imp_df["importance"] / imp_df["importance"].max())**2)
        return imp_df

# second it make one profile data in which feature name , thire quantile according to thier importance and feature value

    def generate_feature_profile(self, top_df):
        imp_df = self.get_importance_quantiles()
        profile = {}

        for _, row in imp_df.iterrows():
            f = row["feature"]
            q = row["quantile"]
            profile[f] = top_df[f].quantile(q)

        return profile

# thier last add thise new strategy into given data

    def create_strategy_row(self, df, strategy_name="My_New_Strategy"):
        top = df.sort_values("Median_score", ascending=False).head(10)
        features = self.generate_feature_profile(top)

        return {
            "Name": strategy_name,
            **features,
            "Wins": None,
            "Median_score": None
        }

    def add_to_dataset(self, df, strategy_name="My_New_Strategy"):
        df.loc[len(df)] = self.create_strategy_row(df, strategy_name)
        return df
```

```python
class titfortat:
    def __init__(self):
        self.last_opponent_move = None

    def move(self, history):
        if history == []:
            return 'C'
        if history[-1] == 'C':
            return 'C'
        return 'D'

```

```python
# decision table
def play_round(a, b):
    if a == 'C' and b == 'C': return (3,3)
    if a == 'D' and b == 'C': return (5,0)
    if a == 'C' and b == 'D': return (0,5)
    return (1,1)
```

```python
# 200 round game and each strategy point

def play_match(strategy1, strategy2, rounds=200):
    h1, h2 = [], []
    s1, s2 = 0, 0

    for _ in range(rounds):
        m1 = strategy1.move(h2)
        m2 = strategy2.move(h1)

        p1, p2 = play_round(m1, m2)

        s1 += p1
        s2 += p2

        h1.append(m1)
        h2.append(m2)

    return s1, s2
    return h1,h2

```

```python
# according to strategy feature behavior make decison


import random

class StrategyFromRow:
    def __init__(self, row):
        self.Initial_C_rate = row["Initial_C_rate"]
        self.CC_to_C_rate   = row["CC_to_C_rate"]
        self.CD_to_C_rate   = row["CD_to_C_rate"]
        self.DC_to_C_rate   = row["DC_to_C_rate"]
        self.DD_to_C_rate   = row["DD_to_C_rate"]
        self.CC_rate   = row["CC_rate"]
        self.CD_rate   = row["CD_rate"]
        self.DC_rate   = row["DC_rate"]
        self.DD_rate   = row["DD_rate"]


    def move(self, history):

        if len(history) == 0:
            return 'C' if random.random() < self.Initial_C_rate else 'D'

        if len(history) == 1:
            last = history[-1]
            if last == 'C':
                return 'C' if random.random() < self.CC_rate else 'D'
            else:
                return 'C' if random.random() < self.DC_rate else 'D'


        if len(history) >= 2:
            last1 = history[-2]
            last0 = history[-1]

            if last1 == 'C' and last0 == 'C':
                p = self.CC_to_C_rate
            elif last1 == 'C' and last0 == 'D':
                p = self.CD_to_C_rate
            elif last1 == 'D' and last0 == 'C':
                p = self.DC_to_C_rate
            else:
                p = self.DD_to_C_rate

            return 'C' if random.random() < p else 'D'



```

```python
def simulate(strategy1, strategy2, runs=50, rounds=200):
    s1_scores, s2_scores = [], []

    for _ in range(runs):
        s1, s2 = play_match(strategy1, strategy2, rounds)
        s1_scores.append(s1)
        s2_scores.append(s2)

    return {
        "new_avg": sum(s1_scores)/runs,
        "top_avg": sum(s2_scores)/runs,
        "new_total": sum(s1_scores),
        "top_total": sum(s2_scores)
    }
```
